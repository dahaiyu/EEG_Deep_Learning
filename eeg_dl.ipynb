```
def assemble_input_output(df, channel, seconds=3):
  all_events = df['event'][0]
  deviation_onset_idxs = []
  reaction_times = []
  for event_idx in range(len(all_events)):
    event = all_events[event_idx]
    if event['type'] == 251.0 or event['type'] == 252.0:
      deviation_onset_idxs.append(event['latency'])
      assert all_events[event_idx+1]['type'] == 253
      reaction_times.append(all_events[event_idx+1]['init_time'] - event['init_time'])

  deviation_onset_idxs = np.array(deviation_onset_idxs)
  reaction_times = np.array(reaction_times)
  median_rt = np.median(reaction_times)
  print('median of', len(all_events)/3, 'reactions is', median_rt, 'seconds')
  binary_rts = np.where(reaction_times > median_rt, 1, 0)

  baselines = []
  for idx in deviation_onset_idxs:
    end_idx = int(idx)
    start_idx = end_idx-500*seconds
    this_baseline = df['data'][channel][start_idx:end_idx]
    baselines.append(this_baseline)
  baselines = np.array(baselines)
  return baselines, binary_rts
```

```
test_baselines, test_binary_rts = assemble_input_output(df, 0)
print(test_baselines.shape)
print(test_binary_rts.shape)
```

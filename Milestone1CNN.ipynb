```
def getAllData(files):

  path_prefix = '/content/drive/Shared drives/CS 230 Project (EEG)/'

  all_events = []
  all_data = []

  for file in files:
    pickle_file = path_prefix + file + ".set.pickle"
    df = pd.read_pickle(pickle_file)
    events = df['event']
    events = np.array(events[0])
    data = df['data']
    data = np.array(data)
    all_events.append(events)
    all_data.append(data)

  return all_events, all_data
```

```
def getCleanChannels(events, data, type_ind, lat_ind, sig_length, sig_max, reactions, median):

  sig_length = sig_length * 500
  signals = []
  reac_times = []

  for i in range(len(events)):
    event = events[i]
    if event[type_ind] == 251 or event[type_ind] == 252:
      index = int(event[lat_ind])
      sig = data[:, index-sig_length:index]
      if np.amax(sig) < sig_max:
        signals.append(sig)
        rt = reactions[int(i/3)]
        reac_times.append(rt)
  
  reac_times = np.array(reac_times)
  return np.array(signals), np.where(reac_times > median, 1, 0)

# all_sigs = []
# all_rts = []
# for i in range(len(all_events)):
#   seconds = 1
#   sig_max = 100
#   signals, reac_times = getCleanChannels(all_events[i], all_data[i], 1, 2, seconds, sig_max, ordered_reacs[i], median)
#   if len(all_sigs) == 0:
#     all_sigs = signals
#     all_rts = reac_times
#   else:
#     all_sigs = np.concatenate((all_sigs, signals))
#     all_rts = np.concatenate((all_rts, reac_times))

# #signals, reac_times = getCleanChannels(events, data, 1, 2, 500, 100, reactions)
# print(np.shape(all_sigs))
# print(np.shape(all_rts))
# print(all_rts)
```
